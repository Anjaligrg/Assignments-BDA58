# MILESTONE 1 DATA COLLECTION AND PREPARATION


```python
# Importing the necessary modules (pandas, numpy, matplotlib.pyplot, seaborn)
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/taxi_tripdata.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
# dropping ehail_fee sice its column is empty or null # Plotting correlation between trip distance and fare amount
df.drop(['ehail_fee'], axis=1, inplace=True)
df
```

```python
df.dtypes
```

```python
# Convert the Id columns to String as best practice (Categorical) :

df[["VendorID",
"RatecodeID",
"PULocationID",
"DOLocationID",
"payment_type"]] = df[["VendorID",
"RatecodeID",
"PULocationID",
"DOLocationID",
"payment_type"]].astype('object')
```

```python
# filling NAN value with mode
for i in df.columns:
    df[i]=df[i].fillna(df[i].mode()[0])
```

```python
#check missing value
missing_values = df.isnull().sum()
missing_values
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

## MILESTONE 2 DATA ANALYSIS AND INSIGHTS


```python
# Display basic statistics
df.describe()
```

```python
# Convert Date's to Pandas Datetime Object :

df['lpep_pickup_datetime'] = pd.to_datetime(df['lpep_pickup_datetime'])

df['lpep_dropoff_datetime'] = pd.to_datetime(df['lpep_dropoff_datetime'])
```

```python
# Extracting the Trip Duration using the Drop Time & Pick Up Time  :
df['trip_duration'] = df['lpep_dropoff_datetime'] - df['lpep_pickup_datetime']
```

```python
# Total Minutes :
df['trip_duration_minutes'] = df['trip_duration'].dt.seconds/60
```

```python
df['tip_percent'] = (df['tip_amount']/df['total_amount']) * 100

df.head()
```

```python
# Categororizing the trip as Morning, Afternoon, Evening, Night based on Pick Up time

def categorize_time_of_day(start_time):
    hour = start_time.hour
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else :
        return 'Night'

df['time_of_day'] = df['lpep_pickup_datetime'].apply(categorize_time_of_day)

df['time_of_day'].value_counts()
```

```python
# Extracting the month from the Pick up time :

df['month'] = df['lpep_pickup_datetime'].dt.strftime('%B')

df['month'].value_counts()
```

```python
# Average Trip distance Calculation :

df['trip_distance'].describe()
```

```python
#average passenger count
df['passenger_count'].value_counts()
```

```python
#Calculate mean tips by passenger_count
mean_tips_by_passenger_count = df.groupby(['passenger_count'])[['tip_amount']].mean()
mean_tips_by_passenger_count
```

# MILESTONE 3 DATA visualization AND REORTING


```python
# Plotting correlation between trip distance and fare amount
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='trip_distance', y='fare_amount', data=df)
plt.title('Correlation between Trip Distance and Fare Amount')
plt.xlabel('Trip Distance (miles)')
plt.ylabel('Fare Amount ($)')
plt.show()
```

```python
# Extracting day of the week from lpep_pickup datetime
df['lpep_pickup_datetime'] = pd.to_datetime(df['lpep_pickup_datetime']) # Removed extra tab character from column name
df['day_of_week'] = df['lpep_pickup_datetime'].dt.day_name() # Removed extra tab character from column name

# Plotting taxi demand by day of the week
plt.figure(figsize=(10, 6))
sns.countplot(x='day_of_week', data=df, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Taxi Demand by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Trips')
plt.show()

```

```python
# Plotting taxi demand by hour of the day
df['hour_of_day'] = df['lpep_pickup_datetime'].dt.hour
plt.figure(figsize=(10, 6))
sns.countplot(x='hour_of_day', data=df)
plt.title('Taxi Demand by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Trips')
plt.show()
```

```python
# Create histogram of tip_amount
plt.figure(figsize=(12,6))
ax = sns.histplot(df['tip_amount'], bins=range(0,21,1))
ax.set_xticks(range(0,21,2))
ax.set_xticklabels(range(0,21,2))
plt.title('Tip amount histogram');
```

```python
# Check the correlation between trip distance, trip duration, fare_amount
# Use another Spearman correlation technique to check the correlation :

sns.heatmap(df[['fare_amount', 'total_amount','trip_distance', 'trip_duration_minutes']].corr(method = 'spearman'))
plt.show()
```
